```python
# Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
import time
# Google Places API Key
gkey = ""
```

```python
# Import the census data into a pandas DataFrame
census_total_pd = pd.read_csv("census_total_2016_geo.csv")

# Preview the data
census_total_pd.head()
```

```python
# Count of unique counties via Id2
# census_total_pd.groupby("Id2").nunique()
```

```python
# Use the lat/lng we recovered to identify schools
for index, row in census_total_pd.iterrows():
    lat = row["Lat"]
    lng = row["Lng"]
                
# Visualize to confirm lat lng appear
# Create an empty column for school count
census_total_pd["School Count"] = ""

# Counter
row_count = 0

# Loop through and run Google search to get all banks in 7 mile radius (11000 meters)
for index, row in census_total_pd.iterrows():
    
    # Create endpoint url using Google Places Radar and the lat/lng we identified earlier
    target_url ="https://maps.googleapis.com/maps/api/place/radarsearch/json?location=%s,%s&radius=11000&keyword=international+school&key=%s" % (census_total_pd.loc[index]["Lat"], census_total_pd.loc[index]["Lng"], gkey)

    
    # This link helps to handily see the JSON generated for each query
    print("Now retrieving county #%s: %s" % (row_count, census_total_pd.loc[index]["Geography"]))
    row_count += 1 
    print(target_url)
    
    # Run a request to grab the JSON at the target URL
    school_data = requests.get(target_url).json()
    
    # Measure school count on the number of results in the retrieved area
    school_count = len(school_data["results"])
    
    print("Final School Count: " + str(school_count))
    print("")
    
    # Store the bank count into the Data Frame
    census_total_pd.set_value(index, "School Count", school_count)
    
    # Reset bank_count (so there is no chance that a previous record is influencing a latter one)
    school_count = 0

# Visualize
census_total_pd.head()
```
