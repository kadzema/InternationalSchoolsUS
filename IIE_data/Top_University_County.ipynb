```python
# Acquire County of each City
for index, row in university_pd.iterrows():
    county = row["County"]

    # Counter
    row_count = 0

    # URL county name
    target_url = "https://maps.googleapis.com/maps/api/geocode/json?address=%s&key=%s" % (row["City"], gkey)
#     print(urllib.parse.urlencode(target_url))

    # Print log to ensure loop is working correctly
    print("Retrieving county name " + str(row_count))
    print(target_url)
    row_count += 1
    
    # Run requests to grab the JSON at the requested URL
    county_name = requests.get(target_url).json()
    
    # Append county names to County column
    # Use try / except to skip any cities with errors
    try:        
        county = county_name["results"][0]["address_components"][1]["long_name"]
        
        university_pd.set_value(index, "County", county)        
    
    except:
        print("Missing field... skipping.")
```
