```python
# Acquire lat and Lng for each City
row_count = 0
for index, row in university_pd.iterrows():
    
    # URL
    target_url = "https://maps.googleapis.com/maps/api/geocode/json?address=%s&key=%s" % (row["City"], gkey)

    
    # Print log
    print("retrieving county lat and lng " + str(row_count))
    print(target_url)
    row_count += 1
    
    # JSON request at URL
    city_location = requests.get(target_url).json()
    
    # Append the lat/lng columns with try / except to QC errors
    try: 
        city_lat = city_location["results"][0]["geometry"]["location"]["lat"]
        city_lng = city_location["results"][0]["geometry"]["location"]["lng"]
        
        university_pd.set_value(index, "Lat", city_lat)
        university_pd.set_value(index, "Lng", city_lng)
        
    except:
        print("Error with city data. Skipping")
        continue
        
# Visualize
university_pd.head()
```
