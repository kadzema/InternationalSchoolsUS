```python
# Import InternationalSchools file as DataFrame
InternatioanlSchools_pd = pd.read_csv("internationalschools_cleaned.csv" ,encoding="iso-8859-1")
InternatioanlSchools_pd.head()
```

```python
# Add columns for lat, lng,
InternatioanlSchools_pd["Lat"] = ""
InternatioanlSchools_pd["Lng"] = ""
InternatioanlSchools_pd.head()
InternatioanlSchools_pd['index'] = range(0, len(InternatioanlSchools_pd))
InternatioanlSchools_pd.head()
```

```python
# Acquire lat and Lng for each City,State
row_count = 0
for index, row in InternatioanlSchools_pd.iterrows():
    
    # URL
    target_url = "https://maps.googleapis.com/maps/api/geocode/json?address=%s,%s&key=%s" % (row["City"],row["State"], gkey)
    url = target_url.replace(' ', '%20')
    
    # Print log
    print("retrieving county lat and lng " + str(row_count))
    print(url)
    row_count += 1
    
    # JSON request at URL
    city_location = requests.get(url).json()
    
    # Append the lat/lng columns with try / except to QC errors
    try: 
        city_lat = city_location["results"][0]["geometry"]["location"]["lat"]
        city_lng = city_location["results"][0]["geometry"]["location"]["lng"]
        
        InternatioanlSchools_pd.set_value(index, "Lat", city_lat)
        InternatioanlSchools_pd.set_value(index, "Lng", city_lng)
        
    except:
        print("Error with city data. Skipping")
        continue
        
# Visualize
InternatioanlSchools_pd.head()
```
