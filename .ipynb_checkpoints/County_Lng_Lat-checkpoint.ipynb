```python
# Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
import time
```

```python
# Google API Key
gkey = ""
```

```python
# Import cities file as DataFrame
counties_pd = pd.read_csv("census_data/counties_Lng_Lat_03122017.csv")
counties_pd.head()
```

```python
# Add columns for lat, lng,
counties_pd["Lat"] = ""
counties_pd["Lng"] = ""
counties_pd.head()
```

```python
# Acquire lat and Lng for each county
row_count = 0
for index, row in counties_pd.iterrows():
    
    # URL
    target_url = "https://maps.googleapis.com/maps/api/geocode/json?address=%s&key=%s" % (row["Geography"], gkey)
    
    
    # Print log
    print("retrieving county lat and lng " + str(row_count))
    print(target_url)
    row_count += 1
    
    # JSON request at URL
    county_location = requests.get(target_url).json()
    
    # Append the lat/lng columns with try / except to QC errors
    try: 
        county_lat = county_location["results"][0]["geometry"]["location"]["lat"]
        county_lng = county_location["results"][0]["geometry"]["location"]["lng"]
        
        counties_pd.set_value(index, "Lat", county_lat)
        counties_pd.set_value(index, "Lng", county_lng)
        
    except:
        print("Error with city data. Skipping")
        continue
        
# Visualize
counties_pd.head()
```
