```python
# Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
import time
```

```python
# Import the census data into a pandas DataFrame
census_hispanic_2000_pd = pd.read_csv("census_data/Hispanic/DEC_00_SF3_P151H_with_ann.csv")
census_hispanic_2000_pd["Year"] = "2000"
census_hispanic_2000_pd["Race"] = "Hispanic"
census_hispanic_2005_pd = pd.read_csv("census_data/Hispanic/ACS_05_EST_B19001I_with_ann.csv")
census_hispanic_2005_pd["Year"] = "2005"
census_hispanic_2005_pd["Race"] = "Hispanic"
census_hispanic_2008_pd = pd.read_csv("census_data/Hispanic/ACS_08_1YR_B19001I_with_ann.csv")
census_hispanic_2008_pd["Year"] = "2008"
census_hispanic_2008_pd["Race"] = "Hispanic"
census_hispanic_2010_pd = pd.read_csv("census_data/Hispanic/ACS_10_1YR_B19001I_with_ann.csv")
census_hispanic_2010_pd["Year"] = "2010"
census_hispanic_2010_pd["Race"] = "Hispanic"
census_hispanic_2013_pd = pd.read_csv("census_data/Hispanic/ACS_13_1YR_B19001I_with_ann.csv")
census_hispanic_2013_pd["Year"] = "2013"
census_hispanic_2013_pd["Race"] = "Hispanic"
census_hispanic_2016_pd = pd.read_csv("census_data/Hispanic/ACS_16_1YR_B19001I_with_ann.csv")
census_hispanic_2016_pd["Year"] = "2016"
census_hispanic_2016_pd["Race"] = "Hispanic"
census_hispanic_2016_pd
```

```python
# Concatenate dataframes]
census_hispanic_total_pd = pd.concat([census_hispanic_2000_pd, census_hispanic_2005_pd,census_hispanic_2008_pd, census_hispanic_2010_pd, census_hispanic_2013_pd, census_hispanic_2016_pd], ignore_index=True)
census_hispanic_total_pd.head() 
```

```python
# Save Data to csv
census_hispanic_total_pd.to_csv("census_data/census_hispanic_total_pd.csv")
```
